```python
def list_files(directory):
    file_list = []
    files_l = []
    for root, directories, files in os.walk(directory):
        for filename in files:
            files_l.append(filename)
            file_list.append(os.path.join(root, filename))
    return files_l

folder_path = "Final_data"
files = list_files(folder_path)
files = [item for item in files if item != ".DS_Store"]
files2 = [file_name.replace('.csv', '') for file_name in files]


print(len(files2))
files2 = sorted(files2)
for file in files2:
    print(file)
```

```python
data_folder = "folder_data"
os.makedirs(data_folder, exist_ok=True)

# Create a folder for each CSV file and move the CSV file into its respective folder
for file_name in files:
    folder_name = file_name.replace('.csv', '')  # Remove the '.csv' extension
    folder_path = os.path.join(data_folder, folder_name)  # Path to the folder in the 'data' directory
    file_name = "Final_data/"+file_name
    os.makedirs(folder_path, exist_ok=True)  # Create the folder in the 'data' directory, if it doesn't exist
    shutil.move(file_name, folder_path)  # Move the CSV file into its respective folder in the 'data' directory

# Check if the folders and files are moved correctly
folders_with_files = {}
for folder in os.listdir(data_folder):
    folder_path = os.path.join(data_folder, folder)
    if os.path.isdir(folder_path):
        files_in_folder = os.listdir(folder_path)
        folders_with_files[folder] = files_in_folder

print("Folders with corresponding CSV files in the 'data' directory:")
for folder, files in folders_with_files.items():
    print(folder + ":")
    print(files)
```
